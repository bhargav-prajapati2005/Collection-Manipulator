##  String Formatting & Manipulation

```python

name = input("Enter student's name: ")
age = int(input("Enter student's age: "))
grade = input("Enter student's grade: ")
subjects = input("Enter subjects (comma separated): "
subject_list = [s.strip().title() for s in subjects.split(",")]
```

```python
from IPython.display import display, Markdown

# Method 1: The Modern f-string (Best for Jupyter)
print("--- F-STRING METHOD ---")
display_text = f"""
Student: {name}
Age: {age}
Grade: {grade}
Enrolled in: {', '.join(subject_list)}
"""
# Using display(Markdown()) makes it look beautiful in Jupyter
display(Markdown(display_text))

print("\n--- .FORMAT() METHOD ---")
header = "Student: {0} | Grade: {1}".format(name, grade)
print(header)
print("Subjects: {}".format(" & ".join(subject_list)))

print("\n--- % OPERATOR METHOD ---")
print("Verification: %s is %d years old." % (name, age))
```

## Mutability and Immutability:


```python
# 1. List of Tuples
# Using Tuples for fixed data: (Student_ID, DOB)
# Using a List to hold multiple student records
student_records = [
    ("S1001", "2005-05-15"),
    ("S1002", "2006-03-22"),
    ("S1003", "2005-11-01"),
    ("S1004", "2006-07-19")
]

# 2. Set for Unique Subjects
# Sets automatically remove duplicates, ensuring a clean list of offerings
subjects_offered = {"Mathematics", "Physics", "English", "History", "Chemistry"}
subjects_offered.add("Mathematics") # This will not add a duplicate

# 3. Dictionary for Detailed Organization
# Keys are Student IDs, Values are Dictionaries containing student details
student_data = {
    "S1001": {
        "name": "Alice Smith",
        "age": 18,
        "grade": "A",
        "subjects": ["Mathematics", "Physics"]
    },
    "S1002": {
        "name": "Bob Johnson",
        "age": 17,
        "grade": "B",
        "subjects": ["English", "History"]
    },
    "S1003": {
        "name": "Charlie Brown",
        "age": 18,
        "grade": "A",
        "subjects": ["Physics", "Chemistry"]
    },
    "S1004": {
        "name": "Diana Prince",
        "age": 17,
        "grade": "A",
        "subjects": ["Mathematics", "History"]
    }
}

# --- Displaying the Results ---

print("--- Unique Subjects Offered ---")
print(subjects_offered)

print("\n--- Student Record Lookup (S1001) ---")
# Accessing the dictionary using a key
record = student_data["S1001"]
print(f"Name: {record['name']} | Grade: {record['grade']}")

print("\n--- Mapping Fixed Data (Tuples) to Detailed Data (Dict) ---")
for student_id, dob in student_records:
    name = student_data[student_id]["name"]
    print(f"ID: {student_id} | Name: {name.ljust(15)} | DOB: {dob}")
```

## Type Casting and del Keyword:


```python
# Example: Casting student data
student_id = int(input("Enter Student ID: "))      # Casts '101' to 101
gpa = float(input("Enter Student GPA: "))        # Casts '3.8' to 3.8
name = input("Enter Student Name: ")              # Remains a string
```

```python
# Our main list of student records
students = [
    {"id": 101, "name": "Alice"},
    {"id": 102, "name": "Bob"},
    {"id": 103, "name": "Charlie"}
]

# 1. Get the ID to delete from the user
target_id = int(input("Enter the Student ID to delete: "))

# 2. Find the index of the student with that ID
index_to_remove = -1

for i in range(len(students)):
    if students[i]["id"] == target_id:
        index_to_remove = i
        break

# 3. Use 'del' to remove the record if found
if index_to_remove != -1:
    del students[index_to_remove]
    print(f"Record with ID {target_id} has been deleted.")
else:
    print("Student ID not found.")

print("Updated list:", students)
```
